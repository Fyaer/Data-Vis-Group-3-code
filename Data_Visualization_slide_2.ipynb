```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Electric_Vehicle_Population_Data (2).csv")

df.columns = df.columns.str.lower().str.replace(" ", "_")

df_filtered = df[(df["model_year"] >= 2005) & (df["model_year"] <= 2024)]

trend_data = df_filtered.groupby(["model_year", "electric_vehicle_type"]).size().reset_index(name="count")

custom_colors = {
    "Battery Electric Vehicle (BEV)": "orange",
    "Plug-in Hybrid Electric Vehicle (PHEV)": "blue"
}

plt.style.use('default')
plt.figure(figsize=(12, 6))

for ev_type in trend_data["electric_vehicle_type"].unique():
    subset = trend_data[trend_data["electric_vehicle_type"] == ev_type]
    plt.plot(subset["model_year"], subset["count"], marker="o",
             label=ev_type, color=custom_colors.get(ev_type, "gray"))

plt.title("Electric Vehicle Adoption Trends by Type", fontsize=16)
plt.xlabel("Model Year", fontsize=12)
plt.ylabel("Number of Registered Vehicles", fontsize=12)
plt.xticks(rotation=45)
plt.legend(title="EV Type")
plt.grid(False)
plt.tight_layout()
plt.show()
```
